```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../inputs/datasets/raw/plant_moniter_health_data.csv')
df.head(3)
```

#### Check for missing values

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
mean = df['Soil_pH'].mean()
standard_deviation = df['Soil_pH'].std()

plt.figure(figsize=(10, 6))
df.plot(kind='hist', bins=50, y='Soil_pH')
plt.axvline(mean + 3 * standard_deviation, color='red', label='+3 std')
plt.axvline(mean - 3 * standard_deviation, color='red', label='-3 std')

plt.show()


df.boxplot(column='Soil_pH', vert=False)
plt.axvline(mean + 3 * standard_deviation, color='red', label='+3 std')
plt.axvline(mean - 3 * standard_deviation, color='red', label='-3 std')

plt.show()
```

#### Examine Humidity

```python
mean = df['Humidity_%'].mean()
standard_deviation = df['Humidity_%'].std()

plt.figure(figsize=(10, 6))
df.plot(kind='hist', bins=50, y='Humidity_%')
plt.axvline(mean + 3 * standard_deviation, color='red', label='+3 std')
plt.axvline(mean - 3 * standard_deviation, color='red', label='-3 std')

plt.show()


df.boxplot(column='Humidity_%', vert=False)
plt.axvline(mean + 3 * standard_deviation, color='red', label='+3 std')
plt.axvline(mean - 3 * standard_deviation, color='red', label='-3 std')

plt.show()
```
